```python
import pandas as pd

#استدعاء ملف البيانات
#استخدم الصف الثاني ك Header بدل الصف الأول المكرر
df = pd.read_csv('/content/Patients transported to hospitals by the SRCA.csv', header=1)
```

```python
df.dtypes
```

```python
df.head()
```

```python
# عدد القيم المفقودة لكل عمود
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
#تحويل جميع الاعمده الرقمية لاعداد صحيحهint
numeric_cols = ['Road accidents', 'Altercation', 'falls', 'Burns', 'Drowning',
                'Other accidents', 'Diseases', 'Total']

df[numeric_cols].replace(',', '', regex=True).astype(int)
```

```python
df.columns
```

```python
#حذف المسافات
df.columns = df.columns.str.strip()
```

```python
# حذف آخر صف
df=df.drop(df.index[-1]).reset_index(drop=True)
df.tail()
```

```python
#حفظ نسخه من البيانات بعد تنظيفها
df.to_csv('cleaned_red_crescent_data.csv', index=False)
```
